```python
import matplotlib.pyplot as plt   # librería para graficar
import numpy as np                # librería para manejo de vectores y utilidades matemáticas

pounds = np.array([3.5, 3.69, 3.44, 3.43, 4.34, 4.42, 2.37])
miles_per_galon = np.array([18, 15, 18, 16, 15, 14, 24])

# configurar el tamaño de grafico
plt.figure(figsize=(8,6))

# crear el grafico
plt.scatter(pounds, miles_per_galon, marker='o', color='blue')

plt.title('Car fuel efficiency in miles per gallon')
plt.xlabel('Pounds in 1000s (feature)')
plt.ylabel('Miles per gallon (label)')

plt.show()
```
